```python
import numpy as np
threshold=1.2
indiv=[]
female_only=[1,30,33,34]
male_only=[31,37]
genders=fam.loc[fam[fam["fid"].astype("int64")<0].shape[0]:].loc[202:]["gender"].values.astype(int)
recalls=[]
aucs=[]
prs=[]
for pheno_id in [48,]:#[0,1,2,16,17,18,28,30,31,33,34,37,40,45,48,49,64]:
    pheno_id=pheno_id
    print( pheno_id,best_69_sex.index.values[pheno_id])
    print(best_69_sex.loc[best_69_sex.index.values[pheno_id]].values)

    if pheno_id in female_only:
        x_test_gender=genders[bed.shape[0]*85//100:]
        female=np.argwhere(x_test_gender.astype('int64')==2)[:,0]
        pred_test=bc_pred.values[:,pheno_id].reshape(-1,1)[female,0]
        loc_y=y_test[female,pheno_id]
        prevalence=len(loc_y[loc_y==1])/len(loc_y)
        fpr, tpr, thresholds = metrics.roc_curve(loc_y,pred_test, pos_label=1)
        aucs.append(metrics.auc(fpr, tpr))
        precision, recall, thresholds = metrics.precision_recall_curve(loc_y, pred_test, pos_label=1)
        prs.append(metrics.auc(recall,precision))
    elif pheno_id in male_only:
        x_test_gender=genders[bed.shape[0]*85//100:]
        male=np.argwhere(x_test_gender.astype('int64')==1)[:,0]
        pred_test=bc_pred.values[:,pheno_id].reshape(-1,1)[male,0]
        loc_y=y_test[male,pheno_id]
        prevalence=len(loc_y[loc_y==1])/len(loc_y)
        fpr, tpr, thresholds = metrics.roc_curve(loc_y,pred_test, pos_label=1)
        aucs.append(metrics.auc(fpr, tpr))
        precision, recall, thresholds = metrics.precision_recall_curve(loc_y, pred_test, pos_label=1)
        prs.append(metrics.auc(recall,precision))
    else:
        pred_test=bc_pred[str(pheno_id)].values
        loc_y=y_test[:,pheno_id]
        prevalence=len(loc_y[loc_y==1])/len(loc_y)
        aucs.append(best_69_sex.loc[best_69_sex.index.values[pheno_id]].values[0])
        prs.append(best_69_sex.loc[best_69_sex.index.values[pheno_id]].values[1])
        true.append([loc_y,pred_test])
        RocCurveDisplay.from_predictions(loc_y,pred_test,name="",color="darkorange")
        plt.plot([0, 1], [0, 1], "k--")
        plt.legend('',frameon=False)
        plt.ylabel("")
        plt.xlabel("")
        PrecisionRecallDisplay.from_predictions(loc_y,pred_test,name="",color="darkorange")
        plt.plot([0, 1], [prevalence,prevalence], "k--")
        plt.legend('',frameon=False)
        plt.ylabel("")
        plt.xlabel("")
    #sort the predicted value and keep indexes
    sorted_indexes=np.argsort(pred_test)#[::-1]
    #sort the true value with the computed indexes
    sorted_y_test=loc_y[sorted_indexes]
    length_tot=len(sorted_indexes)
    odds2=[]
    threshold=2
    individual=0
    # Go over thresholds
    for i in range(0,length_tot):
        #set the threshold
        true_right=sorted_y_test[i:]
        true_left=sorted_y_test[:i]
        #count the cases/control on right threshold (superior PRS) (TP FP)
        ones_right=np.count_nonzero(true_right)
        zeros_right=length_tot-i-ones_right
        #count the cases/control on left threshold (inferior PRS) (FN TN)
        ones_left=np.count_nonzero(true_left)
        zeros_left=i-ones_left
        odd2=ones_right/((ones_right+zeros_right)*prevalence)
        odds2.append(odd2)
        #print(odd2)
        if odd2>threshold:
            print("ODD:"+str(odd2),
                  " Threshold PRS:"+str(pred_test[sorted_indexes][i]),
                  "Individuals remaining:"+str(len(y_test)-individual),
                 " Recall:"+str(ones_right/(ones_right+ones_left)))
            recalls.append(ones_right/(ones_right+ones_left))
            break
        individual+=1
    if i==length_tot-1:
        print("Missed")
        recalls.append(0)
    indiv.append(len(y_test)-individual)
    #plot Density plot with threshold prs
    results=pd.DataFrame(pred_test)
    target=pd.DataFrame(loc_y)
    target.columns=["Prediction Target"]
    result_pred=pd.concat([results,target],axis=1)
    result_pred.columns=["Predicted","Prediction Target"]
    fig, ax = plt.subplots(figsize = (7,10))
    r=result_pred.groupby('Prediction Target')['Predicted']
    r.plot(kind='density',legend=True)
    #ax.set_xlabel("Probabilities=PRS")
    ax.set_ylabel("")
    ax.set_xlim([-0.005, 0.04])
    ax.get_legend().remove()
    plt.xticks(fontsize=13)
    plt.yticks(fontsize=13)
    plt.ticklabel_format(axis='x', style='sci', scilimits=(0,0))
    #print("KS test control:",stats.kstest(result_pred[result_pred["Prediction Target"]==0]["Predicted"].values.reshape(-1, 1),"norm"))
    #print("KS test case:",stats.kstest(result_pred[result_pred["Prediction Target"]==1]["Predicted"].values.reshape(-1, 1),"norm"))
    '''gmm = BayesianGaussianMixture(n_components=4,n_init=100,weight_concentration_prior=10**-7,max_iter=1000,covariance_type="spherical")
    gmm.fit(result_pred[result_pred["Prediction Target"]==1]["Predicted"].values.reshape(-1, 1))
    gmm2 = BayesianGaussianMixture(n_components=4,n_init=100,weight_concentration_prior=10**-7,max_iter=1000,covariance_type="spherical")
    print("2 fit")
    gmm2.fit(result_pred[result_pred["Prediction Target"]==0]["Predicted"].values.reshape(-1, 1))
    laws=gmm.sample(10000)
    lawspd=pd.DataFrame([laws[0][:,0],laws[1]]).T
    print("2 sample")
    laws2=gmm2.sample(10000)
    lawspd2=pd.DataFrame([laws2[0][:,0],laws2[1]]).T
    print("Means:",gmm.means_)
    print("Weights:",gmm.weights_)
    print("Covariance:",gmm.covariances_)
    print('\n')
    print("control distrib")
    print("Means:",gmm2.means_)
    print("Weights:",gmm2.weights_)
    print("Covariance:",gmm2.covariances_)
    '''
    #ax.legend(['    ', '    '])
    fig,ax=plt.subplots()
    ax.plot(odds2)
    #ax.set_title('prevalence_ratio=ones_right/(ones_right+zeros_right)/prevalence')
```
